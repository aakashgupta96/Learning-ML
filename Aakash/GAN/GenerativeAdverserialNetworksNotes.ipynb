# What is GAN

1. Generative Adverserial Networks.
2. GAN networks are used for implementing unsupervised learning algorithms. They are majorly used in the field of images.
3. These network basically consists of 2 major network components.
    1. Generative Networks: They are generally networks of fully connected layers that converts data from one domain to another. For example, converting image of shoes to socks. They can also be called encoders or deconvolutional networks.
    2. Dicriminator Network: There are generally networks of convolution and pooling layers in addition to dense layers that are used for classifying samples to one of the target classes.

# Working of GAN

1. These networks work on the basis of game theory where two players, generator and discriminator plays a game and both are trying to counter-affect each others move.
2. Generator tries to convert random noise or input samples to a domain such that discriminator clssifies it correctly as target class. For ex, if I want to convert random sample to image of shoes, my generator would try to generate image of shoes based on the noise sample such that discriminator also classifies it as image of shoes.
3. On the other hand, discriminator wants to classify images generated by generator for random noises as (not an image of shoes). At the same time, it also wants to correctly classify the actual images of shoes.
4. Thus there is an ongoing competition between generator and discriminator network which eventually leads to an equilibrium such that quality of generator becomes so good that it produces realistic images from random noise or input samples.
2. Generator tries to produce such images that can fool the descriminator.


# Intuition of cost of both components

Let us say, discriminator components realises a function D and generator realises a function G.

For GAN to trasform Z to X:

1. We want dicriminator to easily label X type of input to X and reject Z type of input. So, log probability of D(x type input) should be high and log probability of D(z type of input) should be low. So, its cost function will have 2 components:
    1. One would try to maximize its likelihood to classify x type of input as x ( log(D(x)) should be max)
    2. It should also try to minimize the likehood of D(G(z)) to classify generated object from z (G(z)) as x type (so log(D(G(z)) should be min) or (log(1-D(G(z)) should be max). 
2. We want generator to transform z type of input to x type such that log(D(G(z))) should be high because our generator wants to fool discriminator. Also if G(z) can be similar to x type of input, D(G(z)) would likely classify it as x.
